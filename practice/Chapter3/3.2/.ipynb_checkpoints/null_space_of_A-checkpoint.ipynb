# **Nullspace of a Matrix — $N(A)$**

### **Definition**

The **nullspace** of a matrix $A$ (also called the kernel), denoted $N(A)$, is the set of all vectors $\mathbf{x}$ such that:

$$
A\mathbf{x} = \mathbf{0}
$$

In other words, it is the solution set to the **homogeneous system** of linear equations.

### **Properties**

- $N(A)$ is a **subspace** of $\mathbb{R}^n$ where $n$ is the number of columns of $A$
- It contains **all vectors that get mapped to the zero vector** under $A$
- The **dimension of the nullspace** is called the **nullity** of $A$
- If $A$ is $m \times n$, then:

$$
\text{rank}(A) + \text{nullity}(A) = n
$$

(This is the **Rank-Nullity Theorem**)

### **How to Find the Nullspace:**

1. Solve $A\mathbf{x} = \mathbf{0}$ using **row-reduction (RREF)**
2. Express the solution in **parametric vector form**
3. The vectors that span the nullspace are the **nullspace basis**

### **Code Implementation**

```python
import numpy as np
from sympy import Matrix, init_printing
init_printing()

# Define matrix A (3x4)
A = Matrix([
    [1, 2, -1, 4],
    [2, 4, -2, 8],
    [3, 6, -3, 12]
])

# Step 1: Reduce to RREF
rref_A, pivots = A.rref()
print("RREF of A:")
display(rref_A)

# Step 2: Compute the nullspace
nullspace = A.nullspace()

print("\nBasis for the Nullspace N(A):")
for i, vec in enumerate(nullspace, start=1):
    print(f"Vector {i}:")
    display(vec)
```

### **Visualization**
- For visual purposes, let’s use a $3x3$ matrix with a **1D nullspace** to illustrate the direction collapsed.

```python
import matplotlib.pyplot as plt

# A simpler 3x3 matrix with a 1D nullspace
A = np.array([
    [1, 2, 1],
    [2, 4, 2],
    [3, 6, 3]
])

# We'll visualize a nullspace vector direction
null_vec = np.array([-2, 1, 0])  # known solution to A @ x = 0

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')

# Draw the nullspace line
t = np.linspace(-10, 10, 100)
x = null_vec[0] * t
y = null_vec[1] * t
z = null_vec[2] * t

ax.plot(x, y, z, label="Nullspace direction", color="blue")
ax.quiver(0, 0, 0, null_vec[0], null_vec[1], null_vec[2], color='red', linewidth=2)

ax.set_title("Visualization of Nullspace Direction")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.tight_layout()
plt.show()
```

### **Interpretation**

- Geometrically, the nullspace describes directions in which the matrix **collapses** vectors to zero.
- If $N(A) = \{\mathbf{0}\}$, then $A$ is **injective** (has full column rank).
