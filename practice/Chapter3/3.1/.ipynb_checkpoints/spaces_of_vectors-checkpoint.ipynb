# **Spaces of Vectors**

The **column space** (or **range**) of a matrix $A \in \mathbb{R}^{m \times n}$ is the set of all linear combinations of its column vectors. It answers the question:

> What vectors $\vec{b}$ can be expressed as $A\vec{x} = \vec{b}$?

### Key Concepts:

* The **column space** is a subspace of $\mathbb{R}^m$.
* The **pivot columns** (columns with leading entries in RREF) of a matrix form a **basis** for the column space.
* A vector $\vec{v}$ is **in the column space** of $A$ if the system $A\vec{x} = \vec{v}$ is consistent (has a solution).

We use **Gaussian elimination** to find the pivot columns — these tell us which columns are linearly independent.

```python
import numpy as np
from sympy import Matrix

# Define matrix A
A = Matrix([
    [1, 2, 3],
    [2, 4, 6],
    [1, 1, 2]
])

# Step 1: Show original matrix
display(A)

# Step 2: Perform row reduction
rref_matrix, pivot_cols = A.rref()

print("RREF of A:")
display(rref_matrix)
print("Pivot columns:", pivot_cols)

# Step 3: Extract basis for column space
basis_columns = [A.col(i) for i in pivot_cols]

print("Basis for the column space of A:")
for vec in basis_columns:
    display(vec)
```

* In the matrix above, column 1 and column 3 might be dependent.
* SymPy’s `rref()` helps identify the pivot columns easily.
* The basis for the column space is just the original pivot columns, not the reduced ones.

### **Visualization**
We can visualize the column vectors and the subspace they span using `matplotlib` (only feasible for 2D or 3D).

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Convert sympy vectors to numpy
v1 = np.array(basis_columns[0]).astype(np.float64).flatten()
v2 = np.array(basis_columns[1]).astype(np.float64).flatten()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot origin
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='v1')
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='b', label='v2')

ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])
ax.set_title('Column Space Basis Vectors')
ax.legend()
plt.show()
```
