# **Column Space of a Matrix A $C(A)$**

> The **column space** of a matrix $A$ is the set of all linear combinations of its column vectors. It represents a subspace of $\mathbb{R}^m$, where $m$ is the number of rows in $A$.

### **Definition**

If $A$ is an $m \times n$ matrix with columns $\vec{a}_1, \vec{a}_2, \dots, \vec{a}_n$, then:

$$
C(A) = \text{Span}\{\vec{a}_1, \vec{a}_2, \dots, \vec{a}_n\}
$$

That is, any vector $\vec{b}$ in the column space satisfies:

$$
A\vec{x} = \vec{b}
$$

for some vector $\vec{x} \in \mathbb{R}^n$.

### **Properties**

- $C(A)$ is a **subspace** of $\mathbb{R}^m$
- The **dimension** of $C(A)$ is the number of pivot columns in the reduced row echelon form of $A$
- The **basis** of $C(A)$ is formed by the pivot columns **from the original matrix** (not the reduced one!)

### **Code Implementation**

```python
import numpy as np
from sympy import Matrix

# Define a matrix A
A = Matrix([
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
])

# Display original matrix
display(A)

# Row-reduce A to find pivot columns
rref_matrix, pivot_cols = A.rref()
print("Reduced Row Echelon Form:")
display(rref_matrix)

print(f"Pivot columns (indices): {pivot_cols}")

# Extract basis vectors from original matrix A using pivot column indices
basis_vectors = [A.col(i) for i in pivot_cols]
print("Basis for the column space C(A):")
for vec in basis_vectors:
    display(vec)
```

### **Visualization**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Convert SymPy vectors to NumPy arrays
vecs = [np.array(basis).astype(np.float64).flatten() for basis in basis_vectors]

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot basis vectors
origin = np.zeros(3)
colors = ['r', 'g']
for i, vec in enumerate(vecs):
    ax.quiver(*origin, *vec, color=colors[i], label=f'Basis Vector {i+1}')

# Plot settings
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("Basis Vectors for Column Space C(A)")
ax.legend()
plt.grid()
plt.show()
```

### **Why it Matters**

Understanding $C(A)$ helps answer:

- What vectors $\vec{b}$ can be reached by linear combinations of the columns of $A$?
- Does a system $A\vec{x} = \vec{b}$ have a solution?
