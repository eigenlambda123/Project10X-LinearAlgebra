### 2x2 Matrix Visualization

```python
import matplotlib.pyplot as plt

plt.quiver(0, 0, A[0][0], A[0][1], angles='xy', scale_units='xy', scale=1, color="red")
plt.quiver(0, 0, A[1][0], A[1][1], angles='xy', scale_units='xy', scale=1, color="blue")
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
plt.show()
```

### 3x3 Matrix Visualization

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Define a 3x3 matrix as three column vectors
A = np.array([
    [1, 0, 0],   # First column vector
    [0, 1, 0],   # Second column vector
    [0, 0, 1]    # Third column vector
])  # You can change these to anything, e.g., [[1, 2, 3], [0, 1, 4], [2, 1, 1]]

# Transpose to get column vectors
vectors = B.T

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot each vector as an arrow from the origin
origin = np.zeros(3)
colors = ['r', 'g', 'b']

for vec, color in zip(vectors, colors):
    ax.quiver(
        origin[0], origin[1], origin[2],  # Start at origin
        vec[0], vec[1], vec[2],           # Vector components
        color=color, arrow_length_ratio=0.1
    )

# Set axis limits
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])

# Label axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title("33 Matrix Column Vectors in 3D")
plt.grid(True)
plt.show()
```
