```python
v = [1, 2]
w = [1, 1]

scalar_a = 2
scalar_b = 2
```

### Vector Addition (Column)

```python
import matplotlib.pyplot as plt
import numpy as np

def vector_addition_visual(v1, v2, ans):
    # Plot setup
    plt.figure(figsize=(3, 3)) # cartesian size
    ax = plt.gca()
    ax.set_xlim(-1, 5)
    ax.set_ylim(-1, 5)
    ax.set_aspect('equal')
    plt.grid(True)
    
    # Draw vectors from origin
    plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')
    plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector b (from a)')
    plt.quiver(0, 0, ans[0], ans[1], angles='xy', scale_units='xy', scale=1, color='red', label='a + b')
    
    # Labels
    plt.legend(fontsize=7)
    plt.title("2D Vector Addition")
    plt.xlabel("X", fontsize=5)
    plt.ylabel("Y", fontsize=5)
    
    plt.show()

def add_col(v1, v2):
    res = []
    for i in range(len(v1)):
        res.append(v1[i] + v2[i])
    return res

ans = add_col(v, w)
vector_addition_visual(v, w, ans)
```

### Vector Addition (Row)

```python
def add_row(v1, v2):
    res = []
    for x, y in zip(v1, v2):
        res.append(x + y)
    return res

ans = add_row(v, w)
vector_addition_visual(v, w, ans)
```

### Scalar Multiplication

```python
scalar = 2

def scalar_multiplication_visual(v, scalar, scaled_v):
    
    # Plot setup
    plt.figure()
    ax = plt.gca()
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    ax.set_aspect('equal')
    plt.grid(True)

    # Draw scaled vector
    plt.quiver(0, 0, scaled_v[0], scaled_v[1], angles='xy', scale_units='xy', scale=1, color='red', label=f'{scalar}  vector')

    # Draw original vector
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Original vector')

    # Labels
    plt.legend(fontsize=10)
    plt.title("Scalar Multiplication", fontsize=12)
    plt.xlabel("X", fontsize=10)
    plt.ylabel("Y", fontsize=10)

    plt.show()


def scale(v, scalar):
    res = []
    for i in range(len(v)):
        res.append(v[i] * scalar)
    return res

scaled_v = scale(v, scalar)
scalar_multiplication_visual(v, scalar, scaled_v)
```

### Linear Combinations

```python
def linear_comb_visual(v1, v2, a, b):
    ans = linear_comb(v1, v2, a, b)
    
    # Plot setup
    plt.figure()
    ax = plt.gca()
    ax.set_xlim(-10, 10)
    ax.set_ylim(-10, 10)
    ax.set_aspect('equal')
    plt.grid(True)

    # Base vectors
    plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v1')
    plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='v2')

    # Scaled vectors
    scaled_v1 = [a * v1[0], a * v1[1]]
    scaled_v2 = [b * v2[0], b * v2[1]]

    # Scaled V1
    plt.quiver(0, 0, scaled_v1[0], scaled_v1[1], angles='xy', scale_units='xy', scale=1, color='lightblue', label=f'{a}  v1')
    # Scaled V2
    plt.quiver(0, 0, scaled_v2[0], scaled_v2[1], angles='xy', scale_units='xy', scale=1, color='lightblue', label=f'{b}  v2')
    
    # Scaled V1 + Scaled V2
    plt.quiver(scaled_v1[0], scaled_v1[1], scaled_v2[0], scaled_v2[1], angles='xy', scale_units='xy', scale=1, color='orange', label=f'{b}  v2 (from scaled v1)')

    # Resultant vector
    plt.quiver(0, 0, ans[0], ans[1], angles='xy', scale_units='xy', scale=1, color='green', label='Linear Combination')

    # Labels
    plt.legend(fontsize=9,  loc="lower right")
    plt.title("Linear Combination of Two Vectors", fontsize=12)
    plt.xlabel("X", fontsize=10)
    plt.ylabel("Y", fontsize=10)

    plt.show()

def linear_comb(v1, v2, a, b):
    res = []
    for i in range(len(v1)):
        res.append((v1[i] * a) + (v2[i] * b))
    return res

ans = linear_comb(v, w, scalar_a, scalar_b)
linear_comb_visual(v, w, scalar_a, scalar_b)
```
