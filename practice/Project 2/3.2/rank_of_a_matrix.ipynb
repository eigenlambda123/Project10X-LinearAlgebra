# **The Rank of a Matrix**

### **Definition**

The **rank** of a matrix is the number of **linearly independent rows** or **columns** in the matrix. It is a measure of the **dimension of the row space** or **column space**.

For a matrix $A$:

- $\text{rank}(A)$ = number of **pivot columns** in its **Reduced Row Echelon Form (RREF)**
- $\text{rank}(A)$ tells us how many **dimensions** the linear system spans


### **Intuition**

- Think of rank as the number of **essential directions** a matrix spans.
- If a matrix compresses everything to a line, its rank is 1.
- If it spans the full 2D or 3D space, its rank is 2 or 3.


### **Key Facts**

- $\text{rank}(A) \leq \min(m, n)$ for an $m \times n$ matrix
- $\text{rank}(A) = \text{number of leading 1s in RREF}$
- A matrix is **full rank** if $\text{rank}(A) = \min(m, n)$
- If $\text{rank}(A) < n$, then the columns are **linearly dependent**
- Rank is preserved under **row operations** but not under general matrix multiplication


### **Connection to Solutions of $Ax = b$**

- If $\text{rank}(A) = \text{rank}([A|b])$ → system is **consistent**
- If $\text{rank}(A) = n$ → **unique solution** (when $A$ is square)
- If $\text{rank}(A) < n$ → **infinite solutions** or **no solution**


### **Applications**

- Detecting **linear dependence**
- Understanding **dimension** of image/range space
- Determining **invertibility** (invertible ⇔ full rank square matrix)
- Used in **data compression**, **machine learning**, and **control systems**

### **Code Implementation**

```python
import numpy as np
from sympy import Matrix

# Define matrix A
A = Matrix([
    [1, 2, -1],
    [2, 4, -2],
    [3, 6, -3]
])

# Compute the rank
rank_A = A.rank()

# Display result
print("Matrix A:")
display(A)

print(f"\nRank of A: {rank_A}")
```

### **Visualization**

```python
import matplotlib.pyplot as plt

# 3 rows of A are scalar multiples of [1, 2, -1]
v = np.array([1, 2, -1])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the 3 row vectors (they're collinear)
scalars = [1, 2, 3]
colors = ['r', 'g', 'b']

for i, s in enumerate(scalars):
    vec = s * v
    ax.quiver(0, 0, 0, vec[0], vec[1], vec[2], color=colors[i], label=f"{s} × v")

ax.set_xlim([-4, 4])
ax.set_ylim([-4, 4])
ax.set_zlim([-4, 4])
ax.set_title("Row Vectors of A in ℝ³ (Collinear ⇒ Rank = 1)")
ax.legend()
plt.show()
```
