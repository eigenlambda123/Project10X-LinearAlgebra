# **The Complete Solution to $Ax = b$**

### **Overview**

Solving the linear system $Ax = b$ involves finding all possible vectors $x$ such that when multiplied by matrix $A$, yield the vector $b$.

There are **two components** to the complete solution:

1. **Particular Solution ($x_p$):**
   - A specific solution to $Ax = b$
   - Satisfies the equation exactly: $A x_p = b$

2. **Nullspace Solution ($x_n$):**
   - General solution to the homogeneous equation $Ax = 0$
   - Every solution to $Ax = b$ can be written as:
     $$
     x = x_p + x_n
     $$
   - Where $x_n$ is any linear combination of the nullspace basis vectors

### **Interpretation**

- If the system has **at least one solution**, then:
  $$
  \text{General solution } x = x_p + \text{Null}(A)
  $$

- If **no solution** exists, the system is inconsistent, meaning $b$ is **not in the column space** of $A$.



### **Code Implementation**

```python
from sympy import Matrix, symbols

# Define matrix A and vector b for a consistent system
A = Matrix([
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
])

b = Matrix([6, 15, 24])

# Augmented matrix [A | b]
Ab = A.row_join(b)

# Compute RREF and check consistency
rref_matrix, pivots = Ab.rref()
rank_A = A.rank()
rank_Ab = Ab.rank()

print("RREF of [A | b]:")
display(rref_matrix)

if rank_A != rank_Ab:
    print("System is inconsistent. No solution exists.")
else:
    # Extract particular solution from RREF
    n = A.cols
    x_p = Matrix.zeros(n, 1)  # n x 1 zero vector
    for i, p_col in enumerate(pivots):
        x_p[p_col] = rref_matrix[i, -1]  # Last column value for pivot row

    print("Particular solution x_p:")
    display(x_p)

    # Compute nullspace
    nullspace = A.nullspace()
    print("Basis for the Nullspace N(A):")
    for vec in nullspace:
        display(vec)

    # Form general solution
    k = len(nullspace)
    if k == 0:
        print("Unique solution (no free variables):")
        display(x_p)
    else:
        # Generate symbols for free variables (t0, t1, ...)
        t_sym = symbols('t0:%d' % k)
        x_gen = x_p
        for i in range(k):
            x_gen += t_sym[i] * nullspace[i]
        print("Complete solution x = x_p + c0*v0 + ... :")
        display(x_gen)
```

### **Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import Matrix, symbols
from mpl_toolkits.mplot3d import Axes3D

# Define matrix A and vector b
A = Matrix([
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
])
b = Matrix([6, 15, 24])

# Augmented matrix
Ab = A.row_join(b)

# RREF and particular solution
rref_matrix, pivots = Ab.rref()
n = A.cols
x_p = Matrix.zeros(n, 1)
for i, p_col in enumerate(pivots):
    x_p[p_col] = rref_matrix[i, -1]

# Nullspace basis vectors
nullspace = A.nullspace()
if not nullspace:
    print("Unique solution only — no nullspace to visualize.")
else:
    v = nullspace[0]  # Only 1D nullspace here

    # Convert to NumPy arrays
    x_p_np = np.array(x_p).astype(np.float64).flatten()
    v_np = np.array(v).astype(np.float64).flatten()

    # Create a range of t values
    t_vals = np.linspace(-10, 10, 100)
    sol_points = np.array([x_p_np + t * v_np for t in t_vals])

    # Plotting
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(sol_points[:, 0], sol_points[:, 1], sol_points[:, 2], label='Complete Solution Line', color='blue')

    # Highlight particular solution
    ax.scatter(x_p_np[0], x_p_np[1], x_p_np[2], color='red', s=50, label='Particular Solution $x_p$')

    # Set axis labels
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.set_title('Solution Space for $A\\mathbf{x} = \\mathbf{b}$')
    ax.legend()

    plt.show()
```

### **Summary**

- Reduce augmented matrix $[A \mid b]$ to RREF
- Check if solution exists
- If yes:
  - Find a particular solution $x_p$
  - Find basis of the nullspace of $A$
  - Express the complete solution as:
    $$
    x = x_p + c_1 v_1 + c_2 v_2 + \dots + c_k v_k
    $$
    where $v_1, \dots, v_k$ are nullspace basis vectors
